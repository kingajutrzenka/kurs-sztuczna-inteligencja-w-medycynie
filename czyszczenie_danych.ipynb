## **braki w danych**

```python
df.isnull().sum() #liczba brak√≥w w kolumnach
```

nie mo≈ºemy takich danych u≈ºyƒá do modelu!

```python
import numpy as np
df.replace(['|', 'unknown', 'Unknown', ",", "", ""], np.nan, inplace=True) #wsyztskie dziwne rzeczy kt√≥re siƒô pojawi≈Çy
```

```python
df.isna().sum() #isna = isnull
```

```python
print(df['Age'].unique()) #pokazuje nam dane w tej kolumnie
```

```python
df['Age'] = df["Age"].astype('str').str.replace("Y", '', regex=False)
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
```

 errors='coerce' - to czego nie mo≈ºe przekomwertowaƒá daje nan

```python
print(df)
```

## **usuwanie lub uzupe≈Çnianie brak√≥w**

Modality - typ badania

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
print(df)
```

```python
df["Age"].fillna(40, inplace=True)
print(df)
```

```python
df = df.dropna(subset=['Age', 'Sex']) # usuwa wszystkie wiersze z brakujƒÖcymi danymi w tych kolumnach
print(df)
```

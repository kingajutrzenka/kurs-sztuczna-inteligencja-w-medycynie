```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pydicom
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
import cv2
sns.set()
```

```python
# funkcja wczytywanie obrazów

def load_label_dicom_images(folder, labels_df):
  images = []
  labels = []
  for i in range(len(labels_df)):
    filename = labels_df.iloc[i]['patientId'] + '.dcm' #trzeba patrzeć na nazwy kolumn w pliku
    label = labels_df.iloc[i]['Target']
    filepath = os.path.join(folder, filename)

    try:
      dicom = pydicom.dcmread(filepath)
      img = dicom.pixel_array.astype(np.float32)
      img = (img - img.mean()) / (np.max(img) - np.min(img)) #normalizacja
      img = cv2.resize(img, (128, 128), interpolation = cv2.INTER_CUBIC) #przescalowanie

      images.append(img)
      labels.append(label)
    except:
      print("Pominięto plik:", filename)

  return np.array(images), np.array(labels)
```
